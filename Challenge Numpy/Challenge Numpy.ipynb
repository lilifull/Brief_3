### Modéliser la performance de marathoniens via une régression à l'aide de la bibliothèque Numpy

```python
import pandas as pd
import numpy as np
#Lecture du fichier marathon et stockage des données dans un dataframe
df_marathon = pd.read_table('marathon.txt', header=None)
#Ajout de titre pour les colonnes
df_marathon.columns = ['ville', 'annee', 'temps', 'secondes']
```

```python
df_marathon
```

```python
#Création de la colonne estParis
df_marathon['estParis'] = df_marathon['ville'] == 'PARIS'
```

```python
#Remplacement des valeurs True par 1 et False par 0
df_marathon['estParis'] = df_marathon['estParis'].replace(True, 1).replace(False, 0)
```

```python
df_marathon.head(3)
```

```python
#Création de la colonne estParis 
#Remplacement des valeurs True par 1 et False par 0
df_marathon['estBerlin'] = df_marathon['ville'] == 'BERLIN'
df_marathon['estBerlin'] = df_marathon['estBerlin'].replace(True, 1).replace(False, 0)
```

```python
df_marathon
```

```python
#création des matrices X et Y 
Y = np.matrix(df_marathon[["secondes"]])
X = np.matrix(df_marathon[["annee","estParis","estBerlin"]])
#Application de la formule, .T permet de transposer et .I permet d'avoir l'inverse de la matrice
A = (X.T * X).I * X.T * Y
A
```

```python
# .format permet de remplir les {}, ici avec les éléments de la matrice
print('Les coefficients a0 = {0}, a1 = {1}, a2 = {2}'.format(A[0,0], A[1,0], A[2,0] ))
```
